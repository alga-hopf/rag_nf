# Normal form for right angled groups

These notebook allows to compute explicitly the normal form for elements in a right-angled group (RAG).

```python
#import networkx as nx
#from matplotlib import pyplot as plt
#from itertools import combinations
#from collections import Counter
import os
import sys
#sys.path.insert(0, os.path.abspath('~//home/alice/rag_nf/'))
sys.path.insert(0, os.path.abspath('/home/alice/rag_nf'))
from rag_normal_form import functions_normal_form as nf
```

```python
# Here we define the graph (with the NetworkX sintax) and the sets of black and white vertices
verticesLink={1:[2,3,6],2:[1,3,5],3:[1,2,4],4:[3,5,6],5:[2,4,6],6:[1,4,5]}
Vb=[1,2,4,5] # set of black vertices
G=nf.bic_graph(verticesLink,Vb) # bicolored graph
nf.draw_graph(G) # draw the bicolored graph
```

```python
# How the graph is stored
G
```

```python
# Insert an element of the RAG associated to G in this manner according to the function element
# The first entry is a list of integers varying from -N to N (the number of vertices of the graph) except for 0. 
# The i-th entry of the list is a number n or -n according if the generator x_n or x_n^(-1) is in position i of the word
X=nf.element([3,6,1,4,2,4,3,5],G)
Y=nf.element([6,3,1,3],G)
```

```python
# Print the normal form of X in array format. More precisely, each row corresponds to a left conductor and each column corresponds to a 
# generator. Going from top to bottom, each entry is the power of the corresponding generator in the left conductor.
# Going from left to right anf from top to bottom we recover the normal form of the element
nf.normal_form(X)
```

```python
# Print the normal form of X in terms of the generators x_i. It's possible to choose the variable that represents the generators
nf.print_normal_form('x',X)
```

```python
# Print whether the above elements X and Y are equal
nf.are_equal(X,Y)
```
